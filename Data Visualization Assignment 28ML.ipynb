# In this assignment students have to transform iris data into 3 dimensions and plot a 3d chart with transformed dimensions and colour  each data point with specific class.

### Importing all the libraries which is required to execute our task.

```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
import seaborn as sns
```

```python
from sklearn import datasets
```

```python
df = datasets.load_iris()
```

```python
#importing the data into two different variables
X = df.data[:,:2] 
y= df.target
```

```python
from mpl_toolkits.mplot3d import Axes3D  #library for plotting graph into 3d
from sklearn.decomposition import PCA   #principal component analysis
```

```python
plt.figure(1, figsize=(10, 8))   #1 is axis and figure size as tuple 
# Plot the X,y datasets
plt.scatter(X[:, 0], X[:, 1], c=y, cmap=plt.cm.Set1,  # first two axis & cmap is for color , edgecolor are various param
            edgecolor='k')
plt.xlabel('Sepal length') #we can give label also
plt.ylabel('Sepal width')
```

```python
fig = plt.figure(1, figsize=(14, 10))    #1 i.e axis & we are passing figure size as tuple
ax = Axes3D(fig, elev=-150, azim=110)    #azimuthal viewing angle,numerical value
X_red= PCA(n_components=3).fit_transform(df.data)    #n_components i.e in how many components you want your datasets
ax.scatter(X_red[:, 0], X_red[:, 1], X_red[:, 2], c=y,
           cmap=plt.cm.Set1, edgecolor='k', s=100)
ax.set_title("Decomposed datasets")
ax.set_xlabel("X axis")

ax.set_ylabel("y axis")

ax.set_zlabel("z axis")


plt.show() #command to show our graph
```
